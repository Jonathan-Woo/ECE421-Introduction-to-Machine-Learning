Problem 1 - Clustering with k-means

Question 1

```python
def k_means(data, k):
    '''
    Implementation of the k-means algorithm for data data and k centroids
    Returns cluster centroids and cluster assignment for each data point
    '''
    #get dimension of data
    dim = len(data[0])

    #initialize k centroids using heuristic of randomly choosing 
    #a training point
    centroids = []
    for i in range(k):
        #generate random index of training data
        idx = np.random.randint(len(data))
        #initialize centroid i as selected training data
        centroids.append(data[idx])

    #initialize cluster_assignment list
    cluster_assignment = [0]*len(data)
    
    #loop 50 times
    for _ in range(50):
        #conduct cluster assignment
        #compute euclidean distance of points and keep track of min centroid
        for e, data_point in enumerate(data):
            minDist = 100000
            minCentroid = -1
            for i, centroid in enumerate(centroids):
                dist = np.linalg.norm(centroid - data_point)
                if dist < minDist:
                    minDist = dist
                    minCentroid = i
            cluster_assignment[e] = minCentroid

        #conduct centroid update
        #compute average position for each centroid assignment
        #initiate new array for centroids. Initialize with dummy vector of length dim to retain 2D np shape.
        #We don't need the location of the old centroids anymore
        centroids = [np.array([[0]*dim]) for _ in range(k)]
        for i in range(len(data)):
            centroids[cluster_assignment[i]] = np.append(centroids[cluster_assignment[i]], np.array([data[i]]), axis=0)
        
        #update centroids as mean of assigned data points
        #exclude dummy vector set earlier
        for i in range(len(centroids)):
            centroids[i] = np.mean(centroids[i][1:], axis=0)
    
    return cluster_assignment, centroids
```

```python
def distortion(data, cluster_assignment, centroids):
    '''
    Returns distortion given data, cluster assignments, and centroids
    '''
    #loop through each data point and compute distortion with assigned centroid    
    cur_dist = 0
    for i, data_point in enumerate(data):
        cur_dist += np.linalg.norm(data_point - centroids[cluster_assignment[i]])**2
    return cur_dist/len(data)
```

Question 2

I passed a 569 x 30 2-D NumPy array of data and an integer k indicating the number of centroids to the k-means algorithm. With these parameters, centroids were generated by randomly initialized as training data points. In the following code blocks, k-means is executed 5 times for each k-value to try to optimize centroid initialization.

```python
#load modules
import numpy as np

#download data 
from sklearn.datasets import load_breast_cancer
data = load_breast_cancer()['data']
```

```python
#loop through k values
#since we used the heuristic of randomly choosing training data to initialize centroids, conduct k-means multiple times
#for the same k and store results of centroid assignments and centroids that minimize distortion
results = []

for k in range(2, 8):
    min_distortion = 2**100
    for _ in range(5):
        cur_cluster_assignments, cur_centroids = k_means(data, k)
        cur_distortion = distortion(data, cur_cluster_assignments, cur_centroids)
        if cur_distortion < min_distortion:
            min_cluster_assignments = cur_cluster_assignments
            min_centroids = cur_centroids
            min_distortion = cur_distortion

    results.append([min_cluster_assignments, min_centroids])
```

Question 3

```python
#import libraries
import matplotlib.pyplot as plt

#compute distortion
distortions = []
for cluster_assignment, centroids in results:
    distortions.append(distortion(data, cluster_assignment, centroids))
    
#plot figure
x = [_ for _ in range(2, 8)]
plt.figure()
plt.plot(x, distortions)
plt.title('Distortion vs k-Value')
plt.xlabel('k-Value')
plt.ylabel('Distortion')
```
